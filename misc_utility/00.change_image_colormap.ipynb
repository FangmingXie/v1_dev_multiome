```python
fs = [
    '/u/home/f/f7xiesnm/v1_multiome/cell1_red.png',
    '/u/home/f/f7xiesnm/v1_multiome/cell2_blue.png',
    '/u/home/f/f7xiesnm/v1_multiome/cell3_both.png',
]

# Different colormaps
colormaps = ['Greens', 'Greens', 'Greens']

imgs = []
for f in fs:
    # Load and convert to grayscale
    img = Image.open(f).convert('L')
    img = np.array(img)
    imgs.append(img)

# Apply colormap
fig, axs = plt.subplots(1,3, figsize=(15, 5))
for i, (cmap, img) in enumerate(zip(colormaps, imgs)):
    ax = axs[i]
    ax.imshow(img, cmap=cmap)
    ax.set_title(f'Colormap: {cmap}')
    ax.axis('off')
fig.tight_layout()
plt.show()
```
