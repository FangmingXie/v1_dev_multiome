```python
def plot0(df, frac=0.1):
    fig, axs = plt.subplots(1,4,figsize=(4*4,1*4))
    ax = axs[0]
    sns.histplot(df[0].sample(frac=frac), kde=False, ax=ax)
    ax = axs[1]
    sns.histplot(df[1].sample(frac=frac), kde=False, ax=ax)
    ax = axs[2]
    sns.histplot(df[2].sample(frac=frac), kde=False, ax=ax)
    ax = axs[3]
    sns.histplot(df[3].sample(frac=frac), kde=False, ax=ax)
```

```python
plot0(df1)
plot0(df2)
plot0(df3)
# plot0(df4)
```
