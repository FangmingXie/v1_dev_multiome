```python
df1 = pd.read_csv(f1, header=None) # x y z in micron
df2 = pd.read_csv(f2, header=None) # x y z in micron
df3 = pd.read_csv(f3, header=None) # x y z in micron
df1
```

```python
def plot0(df):
    fig, axs = plt.subplots(1,4,figsize=(4*4,1*4))
    ax = axs[0]
    sns.histplot(df[0], kde=False, ax=ax)
    ax = axs[1]
    sns.histplot(df[1], kde=False, ax=ax)
    ax = axs[2]
    sns.histplot(df[2], kde=False, ax=ax)
    ax = axs[3]
    sns.histplot(df[3], kde=False, ax=ax)
```

```python
plot0(df1)
plot0(df2)
plot0(df3)
```
