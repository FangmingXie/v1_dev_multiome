```python
import numpy as np
import pickle as pkl
import os 
import matplotlib.pyplot as plt
```

```python
pths = [
    'D:\\SWAP\\Vincent\\lt172_exps\\gene\\outputs\\lt172_gene_r1\\registration\\lt172_gene_r1-to-lt172_gene_r2\\aff',
    'D:\\SWAP\\Vincent\\lt172_exps\\gene\\outputs\\lt172_gene_r3\\registration\\lt172_gene_r3-to-lt172_gene_r2\\aff',
    'D:\\SWAP\\Vincent\\lt172_exps\\gene\\outputs\\lt172_gene_r4\\registration\\lt172_gene_r4-to-lt172_gene_r2\\aff',
    'D:\\SWAP\\Vincent\\lt172_exps\\gene\\outputs\\lt172_gene_r5\\registration\\lt172_gene_r5-to-lt172_gene_r2\\aff',
]
```

```python
def get_coords(f):
    """
    """
    with open(f, 'rb') as fh:
        a = pkl.load(fh)
    coords = np.array([i[0] for i in a])
    return coords
```

```python
res = {}
for i, pth in enumerate(pths):
    f = os.path.join(pth, 'fixed_spots.pkl') # r2 spots
    coords = get_coords(f)
    print(coords.shape)
    f = os.path.join(pth, 'moving_spots.pkl') # current spots
    coords_m = get_coords(f)
    print(coords_m.shape)
    res[str(i)+'_f'] = coords
    res[str(i)+'_m'] = coords_m
```

```python
fig, axs = plt.subplots(1,4, figsize=(4*4,4))
for i, ax in enumerate(axs):
    coords   = res[f'{i}_f']
    coords_m = res[f'{i}_m']
    ax.scatter(coords[:,0], coords[:,1], s=5)
    ax.scatter(coords_m[:,0], coords_m[:,1], s=5)
```

```python
fig, axs = plt.subplots(1,4, figsize=(4*4,4))
for i, ax in enumerate(axs):
    coords   = res[f'{i}_m']
    axs[0].scatter(coords[:,0], coords[:,1], s=5)
```
