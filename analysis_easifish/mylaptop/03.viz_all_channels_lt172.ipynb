```python
import os, sys
import numpy as np
import tifffile
import pandas as pd
from PIL import Image
```

```python
%gui qt5
import napari
viewer = napari.Viewer(ndisplay=3)
```

# everthing

```python
names_allrounds = [
    ['Sorcs3', 'Kcnq5', 'Chrm2', 'Adamts2',],
    ['Kcnip3', 'Rorb', 'Cdh13', 'Cntn5',],
    ['Cdh12', 'Gria3', 'Cntnap2', 'Gabrg3',],
    ['Kcnh5', 'RL (Cre)', 'Slc17a7', 'Grm8',],
    ['Ncam2', 'Rfx3', 'Epha10', 'Baz1a',],
]
```

# FISH (before registration)

```python

rounds   = ['r1', 'r2', 'r3', 'r4', 'r5']
channels = ['c0', 'c1', 'c2', 'c4', ]
lwlimits = [ 95, 130,  90, 120,]
uplimits = [110, 250, 170, 180,]
colors   = ['red', 'yellow', 'green', 'blue',]

for r, names in zip(rounds, names_allrounds):
    for cl, c, ll, ul, n in zip(colors, channels, lwlimits, uplimits, names):
        f = f'/Users/fangmingxie/data/easifish/viz_lt172/{r}_{c}_s4.tiff'
        im = tifffile.imread(f)
        vmax = np.percentile(im, 99.9)
        viewer.add_image(np.array(im), 
                         name=f'{r}_{c}_{n}', blending="additive", colormap=cl, contrast_limits=[90,vmax], visible=False)
    
```

```python
f = f'/Users/fangmingxie/data/easifish/viz_lt172/r3_c3_s4.tiff'
im = tifffile.imread(f)
vmax = np.percentile(im, 99)
vmaxmax = np.max(im)
print(vmax, vmaxmax)
viewer.add_image(np.array(im), 
                 name=f'r3_c3_ref', blending="additive", colormap='gray', contrast_limits=[90,vmax], visible=False)
```
